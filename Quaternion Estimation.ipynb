Convention: (w, x, y, z)

```python
def quat_mult(p,q):
    ''' Perform quaternion product q*p'''
    return array([p[0]*q[0] - (p[1]*q[1] + p[2]*q[2] + p[3]*q[3]),
                  p[0]*q[1] + q[0]*p[1] + p[2]*q[3] - p[3]*q[2],
                  p[0]*q[2] + q[0]*p[2] + p[3]*q[1] - p[1]*q[3],
                  p[0]*q[3] + q[0]*p[3] + p[1]*q[2] - p[2]*q[1]])
```

```python
p = array((sqrt(2)/2, 0, sqrt(2)/2, 0)) # A 90' CCW rotation around the y axis i.e. pitch down by 90'
```

```python
p2 = array((sqrt(2)/2, sqrt(2)/2, 0, 0))
```

```python
quat_mult(p2, p)
```

```python
n = 100 # Number of samples
```

```python
sigma = 10
```

```python
norm(randn(3)*sigma)
```

```python
samples = []
for idx in range(n):
    w = randn(3)*sigma
    w0 = norm(w)
    w /= w0
    #print w0
    q = hstack((cos(w0/2), w*sin(w0/2)))
    #print q
    samples.append( quat_mult(p,q))
```

```python
samples
```

```python
pp = zeros((4,4))
for ss in samples:
    pp += outer(ss, ss)
```

```python
pp
```

```python
evals, evecs = eig(pp)
```

```python
evals
```

```python
evecs
```

```python
m = evecs[:,evals.argmax()]
```
